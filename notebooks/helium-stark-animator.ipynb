```python
from heliumstarkanimate import HeliumStarkAnimator
from hsz import HamiltonianMatrix
import numpy as np
%matplotlib notebook
```

```python
n_min = 5
n_max = 6
S = 1
ham = HamiltonianMatrix(n_min=n_min, n_max=n_max, S=S)
animator = HeliumStarkAnimator(ham)
```

```python
state_idx = n_min**2
Efield = np.linspace(0.0, 6.0*10**5, 51) # V/cm
Bfield = 0.01
charge_dists, stark_map = animator.charge_distributions(state_idx, Efield*1e2, Bfield=Bfield)
```

```python
animator.plot(charge_dists[0], figsize=(3,3))
```

```python
animator.save(charge_dists)
```

```python
animator.plot_interactive(Efield, charge_dists, stark_map, state_idx)
```
